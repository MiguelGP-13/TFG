## ACA DATA

Pares de frases en 2 lenguas.

Buscaremos pares con alguna de las lenguas que nos interesan

```python
import pandas as pd

from datasets import load_dataset

ds = load_dataset("BSC-LT/ACAData", split="train")
```

```python
ACA = ds.to_pandas()
ACA.head()
```

```python
pd.concat((ACA.lang1_code, ACA.lang2_code),axis="rows").value_counts()[["es", "ca", "eu", "gl", "ast"]]
```

Vemos que tiene bastante información de catalán (+120k filas), menos del euskera (+11k filas) y el gallego (+11k filas) y muy pocas del asturiano (45 filas)
